# June Gloom — first look

Pull hourly cloud cover for a few SoCal stations and quantify how gloomy June really is.

Run from the repo root with the venv active. Requires network access for the Open-Meteo API.

```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent / 'src'))

import pandas as pd
from june_gloom.fetch import fetch_june, STATIONS
from june_gloom import analyze, plots

list(STATIONS)
```

```python
# Fetch June 2023 for the coast vs. inland
frames = [fetch_june(st, 2023) for st in ['santa_monica', 'san_diego', 'riverside_inland']]
df = pd.concat(frames, ignore_index=True)
df.head()
```

```python
# Daily gloom summary + count of gloomy mornings
daily = analyze.daily_gloom_summary(df)
analyze.june_gloom_by_year(daily)
```

```python
# The burn-off curve: mean cloud cover by hour of day
plots.plot_hourly_cloud_profile(df);
```
